```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Para sabermos a classifica√ß√£o final do campeonato 
df=pd.read_csv("classifica√ßao_intervalo.csv", header=0)
df.head(18)
```

```python
#Qual foi o p√≥dio deste campeonato nacional caso contassem apenas os jogos at√© ao intevalo ?
podio = df[['equipa', 'pontos']]
podio.head(3)
```

```python
#Quem seria o campe√£o deste campeonato caso contassem apenas os jogos at√© ao intevalo? 
#Com quantas V/E/D, pontos e golos sofridos e marcados?
df.head(1)
```

```python
#Qual √© a m√©dia de pontos ganhos em jogos at√© o intervalo pelas equipas por cada jornada?

pd.options.display.float_format = '{:,.3f}'.format
avg_pontos = df["pontos"] / 34
avg_pontos
df["avg_pontos"] = df["pontos"] /34
result = df[["equipa", "avg_pontos"]]
result
```

```python
#Qual √© a m√©dia de golos marcados por jogo ao intervalo pelas equipas ao longo do campeonato?

pd.options.display.float_format = '{:,.3f}'.format
avg_golos_marcados = df["golos_marcados"] / 34
avg_golos_marcados
df["avg_golos_marcados"] = df["golos_marcados"] /34
result = df[["equipa", "avg_golos_marcados"]]
result
```

```python
#Qual √© a m√©dia de golos sofridos por jogo ao intervalo pelas equipas ao longo do campeonato

pd.options.display.float_format = '{:,.3f}'.format
avg_golos_sofridos = df["golos_sofridos"] / 34
avg_golos_sofridos
df["avg_golos_sofridos"] = df["golos_sofridos"] /34
result = df[["equipa", "avg_golos_sofridos"]]
result
```

```python
#Qual foi a equipa que mais golos marcou ao intervalo (melhor ataque) nesta edi√ß√£o do campeonato?

print(df['golos_marcados'].max())
golos_m = df['golos_marcados'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa que mais golos sofreu ao intervalo (pior defesa) nesta edi√ß√£o do campeonato?

print(df['golos_sofridos'].max())
golos_m = df['golos_sofridos'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa com o maior n√∫mero de vit√≥rias ao intervalo nesta edi√ß√£o do campeonato?

print(df['vitorias'].max())
golos_m = df['vitorias'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa com o maior n√∫mero de empates ao intervalo nesta edi√ß√£o do campeonato?

print(df['empates'].max())
golos_m = df['empates'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa com o maior n√∫mero de derrotas ao intervalo nesta edi√ß√£o do campeonato?

print(df['derrotas'].max())
golos_m = df['derrotas'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Top 5 de equipas com mais golos marcados ao intervalo nesta edi√ß√£o do campeonato

df.sort_values(by = ["golos_marcados"], inplace =False, ascending=False).head()
```

```python
#Top 5 de equipas com mais golos sofridos ao intervalo nesta edi√ß√£o do campeonato

df.sort_values(by = ["golos_sofridos"], inplace =False, ascending=False).head()
```

```python
#Gr√°fico de golos totais marcados ao intervalo pelas equipas (casa e fora) ao longo do campeonato

_ = plt.figure(figsize=(10,8))
_ = plt.bar(df["equipa"], df["golos_marcados"],align = "center", color=("blue"))
_ = plt.xlabel("Equipas")
_ = plt.ylabel("Golos Marcados")
_ = plt.xticks(rotation = 90)
_ = plt.title("Golos marcados ao intervalo na Liga Portugal 21-22")
plt.show()
```

```python
#Gr√°fico de golos totais sofridos (casa e fora) pelas equipas ao longo do campeonato

_ = plt.figure(figsize=(10,8))
_ = plt.bar(df["equipa"], df["golos_sofridos"],align = "center", color=("blue"))
_ = plt.xlabel("Equipas")
_ = plt.ylabel("Golos Sofridos")
_ = plt.xticks(rotation = 90)
_ = plt.title("Golos sofridos ao intervalo na Liga Portugal 21-22")
plt.show()
```

```python
#Gr√°fico da pontua√ß√£o final ao intervalo do campeonato 

_ = plt.figure(figsize=(10,8))
_ = plt.bar(df["equipa"], df["pontos"],align = "center", color=("blue"))
_ = plt.xlabel("Equipas")
_ = plt.ylabel("Pontos")
_ = plt.xticks(rotation = 90)
_ = plt.title("Pontua√ß√£o final ao intervalo da Liga Portugal 21-22")
plt.show()
```
