```python
import pandas as pd
import matplotlib.pyplot as plt

#Para sabermos a classificação final dos guarda redes com mais defesas
df=pd.read_csv("defesas_guarda_redes.csv", header=0)
df.head(20)
```

```python
#Quem foi o guarda-redes com mais defesas deste campeonato? Onde joga e quantas defesas fez ao longo do campeonato?
df.head(1)
```

```python
#Quantidade de guarda-redes no top 20 de guarda-redes com mais defesas da liga (onde jogam)
jogador_clube = df["clube"].value_counts()
jogador_clube
```

```python
#Média de defesas em cada jogo pelos guarda-redes no top 20 de melhores marcadores

pd.options.display.float_format = '{:,.3f}'.format
avg_defesas = df["numero_defesas"] / 34
avg_defesas
pd.options.display.float_format = '{:,.3f}'.format
avg_defesas_jogo = df["numero_defesas"] / 34
avg_defesas_jogo
df["avg_defesas_jogo"] = df["numero_defesas"] /34
result = df[["nome_guarda_redes", "avg_defesas_jogo"]]
result
```

```python
#Qual foi o pódio de guarda-redes com mais defesas deste campeonato nacional?
df.head(3)
```

```python
#Gráfico dos guarda-redes com mais defesas deste campeonato

_ = plt.figure(figsize=(10,8))
_ = plt.bar(df["nome_guarda_redes"], df["numero_defesas"],align = "center", color=("black"))
_ = plt.xlabel("Jogadores")
_ = plt.ylabel("Nº Defesas")
_ = plt.xticks(rotation = 90)
_ = plt.title("Guarda-Redes com mais defesas da Liga Portugal 21-22")
plt.show()
```
