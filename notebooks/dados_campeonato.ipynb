```python
import pandas as pd
import matplotlib.pyplot as plt

#Para sabermos a classifica√ß√£o final do campeonato 
df=pd.read_csv("classifica√ßao.csv", header=0)
df.head(18)
```

```python
#Qual foi o p√≥dio deste campeonato nacional?
podio = df[['equipa', 'pontos']]
podio.head(3)
```

```python
#Quem foi o campe√£o deste campeonato? Com quantas V/E/D, pontos e golos sofridos e marcados?
df.head(1)
```

```python
#Qual √© a m√©dia de pontos ganhos pelas equipas por cada jornada?

pd.options.display.float_format = '{:,.3f}'.format
avg_pontos = df["pontos"] / 34
avg_pontos
df["avg_pontos"] = df["pontos"] /34
result = df[["equipa", "avg_pontos"]]
result
```

```python
#Qual √© a m√©dia de golos marcados por jogo pelas equipas ao longo do campeonato

pd.options.display.float_format = '{:,.3f}'.format
avg_golos_marcados = df["golos_marcados"] / 34
avg_golos_marcados
df["avg_golos_marcados"] = df["golos_marcados"] /34
result = df[["equipa", "avg_golos_marcados"]]
result
```

```python
#Qual √© a m√©dia de golos sofridos por jogo pelas equipas ao longo do campeonato

pd.options.display.float_format = '{:,.3f}'.format
avg_golos_sofridos = df["golos_sofridos"] / 34
avg_golos_sofridos
df["avg_golos_sofridos"] = df["golos_sofridos"] /34
result = df[["equipa", "avg_golos_sofridos"]]
result
```

```python
#Qual foi a equipa que mais golos marcou (melhor ataque) nesta edi√ß√£o do campeonato?

print(df['golos_marcados'].max())
golos_m = df['golos_marcados'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa que mais golos sofreu (pior defesa) nesta edi√ß√£o do campeonato?

print(df['golos_sofridos'].max())
golos_m = df['golos_sofridos'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa com o maior n√∫mero de vit√≥rias nesta edi√ß√£o do campeonato?
print(df['vitorias'].max())
golos_m = df['vitorias'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa com o maior n√∫mero de empates nesta edi√ß√£o do campeonato?
print(df['empates'].max())
golos_m = df['empates'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa com o maior n√∫mero de derrotas nesta edi√ß√£o do campeonato?
print(df['derrotas'].max())
golos_m = df['derrotas'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Top 5 de equipas com mais golos marcados

df.sort_values(by = ["golos_marcados"], inplace =False, ascending=False).head()
```

```python
#Top 5 de equipas com mais golos sofridos

df.sort_values(by = ["golos_sofridos"], inplace =False, ascending=False).head()
```

```python
#Gr√°fico de golos marcados pelas equipas (casa e fora) ao longo do campeonato
_ = plt.figure(figsize=(10,8))
_ = plt.bar(df["equipa"], df["golos_marcados"],align = "center", color=("green"))
_ = plt.xlabel("Equipas")
_ = plt.ylabel("Golos Marcados")
_ = plt.xticks(rotation = 90)
_ = plt.title("Golos marcados na Liga Portugal 21-22")
plt.show()
```

```python
#Gr√°fico de golos sofridos (casa e fora) pelas equipas ao longo do campeonato
_ = plt.figure(figsize=(10,8))
_ = plt.bar(df["equipa"], df["golos_sofridos"],align = "center", color=("red"))
_ = plt.xlabel("Equipas")
_ = plt.ylabel("Golos Sofridos")
_ = plt.xticks(rotation = 90)
_ = plt.title("Golos sofridos na Liga Portugal 21-22")
plt.show()
```

```python
#Gr√°fico da pontua√ß√£o final do campeonato 

_ = plt.figure(figsize=(10,8))
_ = plt.bar(df["equipa"], df["pontos"],align = "center", color=("black"))
_ = plt.xlabel("Equipas")
_ = plt.ylabel("Pontos")
_ = plt.xticks(rotation = 90)
_ = plt.title("Pontua√ß√£o final da Liga Portugal 21-22")
plt.show()
```
