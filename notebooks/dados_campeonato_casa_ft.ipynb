```python
import pandas as pd
import matplotlib.pyplot as plt

#Para sabermos a classifica√ß√£o final do campeonato (apenas contando os jogos em casa)
df=pd.read_csv("classifica√ßao_casa.csv", header=0)
df.head(18)
```

```python
#Qual o p√≥dio deste campeonato se apenas forem contabilizados os jogos em casa?

podio = df[['equipa', 'pontos']]
podio.head(3)
```

```python
#Qual foi a melhor equipa a jogar em casa este campeonato? Com quantas V/E/D, pontos e golos sofridos e marcados?

df.head(1)
```

```python
#Qual √© a m√©dia de pontos ganhos em casa pelas equipas por cada jornada?

pd.options.display.float_format = '{:,.3f}'.format
avg_pontos = df["pontos"] / 17
avg_pontos
df["avg_pontos"] = df["pontos"] /17
result = df[["equipa", "avg_pontos"]]
result
```

```python
#Qual √© a m√©dia de golos marcados em jogos caseiros pelas equipas ao longo do campeonato

pd.options.display.float_format = '{:,.3f}'.format
avg_golos_marcados = df["golos_marcados"] / 17
avg_golos_marcados
df["avg_golos_marcados"] = df["golos_marcados"] /17
result = df[["equipa", "avg_golos_marcados"]]
result
```

```python
#Qual √© a m√©dia de golos sofridos em jogos caseiros pelas equipas ao longo do campeonato

pd.options.display.float_format = '{:,.3f}'.format
avg_golos_sofridos = df["golos_sofridos"] / 17
avg_golos_sofridos
df["avg_golos_sofridos"] = df["golos_sofridos"] /17
result = df[["equipa", "avg_golos_sofridos"]]
result
```

```python
#Qual foi a equipa que mais golos marcou em casa (melhor ataque caseiro) nesta edi√ß√£o do campeonato?

print(df['golos_marcados'].max())
golos_m = df['golos_marcados'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa que mais golos sofreu em casa (pior defesa caseira) nesta edi√ß√£o do campeonato?

print(df['golos_sofridos'].max())
golos_m = df['golos_sofridos'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa com o maior n√∫mero de vit√≥rias caseiras nesta edi√ß√£o do campeonato?
print(df['vitorias'].max())
golos_m = df['vitorias'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa com o maior n√∫mero de empates caseiros nesta edi√ß√£o do campeonato?
print(df['empates'].max())
golos_m = df['empates'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa com o maior n√∫mero de derrotas caseiras nesta edi√ß√£o do campeonato?
print(df['derrotas'].max())
golos_m = df['derrotas'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Top 5 de equipas com mais golos marcados em casa

df.sort_values(by = ["golos_marcados"], inplace =False, ascending=False).head()
```

```python
#Top 5 de equipas com mais golos sofridos em casa

df.sort_values(by = ["golos_sofridos"], inplace =False, ascending=False).head()
```

```python
#Gr√°fico de golos marcados em casa pelas equipas ao longo do campeonato

_ = plt.figure(figsize=(10,8))
_ = plt.bar(df["equipa"], df["golos_marcados"],align = "center", color=("green"))
_ = plt.xlabel("Equipas")
_ = plt.ylabel("Golos Marcados")
_ = plt.xticks(rotation = 90)
_ = plt.title("Golos marcados em casa na Liga Portugal 21-22")
plt.show()
```

```python
#Gr√°fico de golos sofridos em casa pelas equipas ao longo do campeonato

_ = plt.figure(figsize=(10,8))
_ = plt.bar(df["equipa"], df["golos_sofridos"],align = "center", color=("red"))
_ = plt.xlabel("Equipas")
_ = plt.ylabel("Golos Sofridos")
_ = plt.xticks(rotation = 90)
_ = plt.title("Golos sofridos em casa na Liga Portugal 21-22")
plt.show()
```

```python
#Gr√°fico da pontua√ß√£o final (apenas contabilizando os jogos em casa) do campeonato 

_ = plt.figure(figsize=(10,8))
_ = plt.bar(df["equipa"], df["pontos"],align = "center", color=("black"))
_ = plt.xlabel("Equipas")
_ = plt.ylabel("Pontos")
_ = plt.xticks(rotation = 90)
_ = plt.title("Pontua√ß√£o final (apenas contabilizando os jogos em casa) da Liga Portugal 21-22")
plt.show()
```
