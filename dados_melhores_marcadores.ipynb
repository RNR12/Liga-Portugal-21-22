```python
import pandas as pd
import matplotlib.pyplot as plt

#Para sabermos a classificação final dos marcadores do campeonato
df=pd.read_csv("melhores_marcadores.csv", header=0)
df.head(20)
```

```python
#Quem foi o campeão deste campeonato? Em que clube joga o jogador e quantos golos marcou?
df.head(1)
```

```python
#Número de jogadores das equipas no top 20 de melhores marcadores da liga
jogador_clube = df["clube"].value_counts()
jogador_clube
```

```python
#Média de golos marcados em cada jogo pelos jogadores no top 20 de melhores marcadores

pd.options.display.float_format = '{:,.3f}'.format
avg_golos_marcados = df["numero_golos"] / 34
avg_golos_marcados
pd.options.display.float_format = '{:,.3f}'.format
avg_golos_marcados_jogador = df["numero_golos"] / 34
avg_golos_marcados_jogador
df["avg_golos_marcados_jogador"] = df["numero_golos"] /34
result = df[["nome_jogador", "avg_golos_marcados_jogador"]]
result
```

```python
#Qual foi o pódio de marcadores deste campeonato nacional?
df.head(3)
```

```python
#Gráfico dos melhores marcadores do campeonato, separado por cores da equipa
#Red- Benfica; Blue- FC Porto; #910000- SC Braga; #309100- Sporting CP; #dbdbdb- V. Guimarães; #00a5c9- Famalicão
# #d10000- Gil Vicente; Black- Boavista FC; Yellow- Estoril; #084d00- Marítimo; #d6d613- FC Arouca; #116ac2- FC Vizela
# #f6ff73- Tondela; 

_ = plt.figure(figsize=(10,8))
_ = plt.bar(df["nome_jogador"], df["numero_golos"],align = "center", color=("red", "blue", "#910000", "#d10000", "#309100", "#dbdbdb", "blue", "blue", "#00a5c9", "#d10000", "black", "#309100", "yellow", "#309100", "#084d00", "#d6d613", "#116ac2", "#f6ff73", "#309100", "red"))
_ = plt.xlabel("Jogadores")
_ = plt.ylabel("Nº golos marcados")
_ = plt.xticks(rotation = 90)
_ = plt.title("Melhores Marcadores Liga Portugal 21-22")
plt.show()
```
