# Dados estatísticos de jogos caseiros do campeonato decorridos até ao intervalo

```python
import pandas as pd
import matplotlib.pyplot as plt

#Para sabermos a classificação final do campeonato em casa até ao intervalo
df=pd.read_csv("classificaçao_intervalo_casa.csv", header=0)
df.head(18)
```

```python
#Qual o pódio deste campeonato se apenas forem contabilizados os jogos em casa até ao intervalo?

podio = df[['equipa', 'pontos']]
podio.head(3)
```

```python
#Qual foi a melhor equipa a jogar em casa este campeonato, apenas contabilizando as primeiras partes? 
#Com quantas V/E/D, pontos e golos sofridos e marcados?
df.head(1)
```

```python
#Até ao final do primeiro tempo, qual é a média de pontos ganhos pelas equipas a jogar em casa por cada jornada?

pd.options.display.float_format = '{:,.3f}'.format
avg_pontos = df["pontos"] / 17
avg_pontos
df["avg_pontos"] = df["pontos"] /17
result = df[["equipa", "avg_pontos"]]
result
```

```python
#Até ao final do primeiro tempo, qual é a média de golos marcados pelas equipas a jogarem em casa ao longo do campeonato?

pd.options.display.float_format = '{:,.3f}'.format
avg_golos_marcados = df["golos_marcados"] / 17
avg_golos_marcados
df["avg_golos_marcados"] = df["golos_marcados"] /17
result = df[["equipa", "avg_golos_marcados"]]
result
```

```python
#Até ao final do primeiro tempo, qual é a média de golos sofridos pelas equipas a jogarem em casa ao longo do campeonato?

pd.options.display.float_format = '{:,.3f}'.format
avg_golos_sofridos = df["golos_sofridos"] / 17
avg_golos_sofridos
df["avg_golos_sofridos"] = df["golos_sofridos"] /17
result = df[["equipa", "avg_golos_sofridos"]]
result
```

```python
#Qual foi a equipa que mais golos marcou em casa durante a 1ª parte (melhor ataque caseiro ao intervalo) nesta edição do campeonato?

print(df['golos_marcados'].max())
golos_m = df['golos_marcados'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa que mais golos sofreu golos em casa durante a 1ª parte (pior defesa caseira ao intervalo) nesta edição do campeonato?

print(df['golos_sofridos'].max())
golos_m = df['golos_sofridos'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa com o maior número de vitórias caseiras ao intervalo nesta edição do campeonato?
print(df['vitorias'].max())
golos_m = df['vitorias'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa com o maior número de empates caseiros ao intervalo nesta edição do campeonato?
print(df['empates'].max())
golos_m = df['empates'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Qual foi a equipa com o maior número de derrotas caseiras ao intervalo nesta edição do campeonato?
print(df['derrotas'].max())
golos_m = df['derrotas'].idxmax()
print(df['equipa'][golos_m])
```

```python
#Top 5 de equipas com mais golos marcados em casa até ao intervalo

df.sort_values(by = ["golos_marcados"], inplace =False, ascending=False).head()
```

```python
#Top 5 de equipas com mais golos sofridos em casa até o intervalo

df.sort_values(by = ["golos_sofridos"], inplace =False, ascending=False).head()
```

```python
#Gráfico de golos marcados pelas equipas a jogar em casa e até o intervalo ao longo do campeonato

_ = plt.figure(figsize=(10,8))
_ = plt.bar(df["equipa"], df["golos_marcados"],align = "center", color=("green"))
_ = plt.xlabel("Equipas")
_ = plt.ylabel("Golos Marcados")
_ = plt.xticks(rotation = 90)
_ = plt.title("Golos marcados (em casa e até o intervalo) na Liga Portugal 21-22")
plt.show()
```

```python
#Gráfico de golos sofridos (em casa e até o intervalo) pelas equipas ao longo do campeonato

_ = plt.figure(figsize=(10,8))
_ = plt.bar(df["equipa"], df["golos_sofridos"],align = "center", color=("red"))
_ = plt.xlabel("Equipas")
_ = plt.ylabel("Golos Sofridos")
_ = plt.xticks(rotation = 90)
_ = plt.title("Golos sofridos (em casa e até o intervalo) na Liga Portugal 21-22")
plt.show()
```

```python
#Gráfico da pontuação final (apenas contabilizando jogos até ao intervalo e em casa das equipas) do campeonato 

_ = plt.figure(figsize=(10,8))
_ = plt.bar(df["equipa"], df["pontos"],align = "center", color=("black"))
_ = plt.xlabel("Equipas")
_ = plt.ylabel("Pontos")
_ = plt.xticks(rotation = 90)
_ = plt.title("Pontuação final (apenas contando jogos em casa e até o intervalo) da Liga Portugal 21-22")
plt.show()
```
